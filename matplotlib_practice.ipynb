```python
# import modules
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
y2_axis = [12.02, 27.24, 44.20, 50.42, 22.34, 37.04, 53.82, 21.56, 19.85, 33.90, 21.71, 40.09]
```

```python
# plot me
plt.plot(x_axis, y_axis, y2_axis)
```

```python
# now with objects
fig, ax = plt.subplots(figsize=(20, 10))
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_ylim(0,60)
ax.set_title("PyBer Fare by Month")
ax.grid()
ax.plot(x_axis, y_axis, marker='D', color="green", linewidth=2,
        label="Boston")
ax.plot(x_axis, y2_axis, marker='*', color="Blue", linewidth=4,
        label="NYC")
ax.legend()
```

```python
# Create the plot with alternate means of assigning the obj from the list plt.subplots()
# these create an empty chart
fig = plt.figure()
ax = fig.add_subplot()
# this fills it
ax.plot(x_axis, y_axis)
```

```python
# and one more 'fast' option
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
# plt.plot(x_axis, y_axis, label='Boston')
# take 2 - make it a bit more interesting
#plt.figure(figsize=(20, 10))
plt.subplots(figsize=(20, 10))
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
#plt.xlim("Jan", "June")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
plt.grid(axis='y')
```

# Scatter & Buble Plots
### MATLIB style first

```python
# using plot
plt.plot(x_axis, y_axis, 'o')
```

```python
# using scatter
plt.scatter(y_axis, x_axis, c='red', label='Chicago')
plt.title("PyBEr Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.gca().invert_yaxis()
plt.legend()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
#bubbles
#plt.scatter(x_axis, y_axis, s=y_axis_larger)
# now use the list comprehension 
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

# Scatter & Buble Plots
### OO style now

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c='skyblue', alpha=.8, edgecolors='b', linewidth = 2, 
           s= [i * 5 for i in y_axis], label = 'Boston')
ax.invert_yaxis()
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Date")
ax.set_xlabel('Fare ($)')
ax.legend()
plt.show()
```

# Pie Charts
### MatPlot style

```python
# pie
plt.subplots(figsize=(15,12))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, labels=x_axis, colors=colors, explode=explode_values, autopct='%.1f%%')
plt.show()
```

# Pie Charts
### OO style

```python
fig, ax = plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "pink", "tomato", "mistyrose"]
ax.pie(y_axis, labels=x_axis, startangle=-95, colors=colors, explode=explode_values, autopct='%.1f%%')
ax.invert_yaxis()
plt.show()
```
